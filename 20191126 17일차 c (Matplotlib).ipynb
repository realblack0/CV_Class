### axes
- axes는 그래프 이다.
- figure는 그래프를 넣을 가상의 공간이다.
    - figure 안에 여러 개의 그래프를 넣을 수 있다.

```python
# 4-tuple of floats *rect* = [left, bottom, width, height]
plt.axes([0.5, 0, .5, 0])
```

> - height가 0이라서 안보이는 거임!

```python
# 4-tuple of floats *rect* = [left, bottom, width, height]
plt.axes([0, 0, .5, .5]);
```

```python
plt.axes([0.5, 0, .5, 0])
plt.axes([0, 0, .5, .5]);
```

> - axes를 두개 쓰면 그래프가 2개!

```python
plt.axes([0.5, 0, .5, 0.5])
plt.axes([0, 0, .5, .5]);
```

> - 그래프를 여러개 그릴때 사용할 수 있다.

```python
plt.axes([0.5, 0, .5, 0.5])
plt.title("1")
plt.axes([0, 0, .5, .5])
plt.title("2");
```

> - 각 그래프마다 다른 값 쓸 수 있다.

### subplots
- 그래프를 여러개 그리는 기법
- 객체지향 방식

```python
fig, ax = plt.subplots(1,2)
```

> - 객체 지향 방식
- 할당한다. (state model과 다름)

```python
fig, ax = plt.subplots(1,2)
print(ax)
```

> - ax는 그래프이다.
- subplot 2개라서 값이 2개 들어있다.

```python
fig, ax = plt.subplots(1,2)
print(ax[0])
```

```python
fig, ax = plt.subplots(1,2)
print(ax[1])
```

> - ax의 element 각각이 객체이다.

```python
fig, ax = plt.subplots(1,2)
ax[0].set_title("1")
```

> - 객체지향 방식으로 `.`으로 접근해서 메스드를 쓸 수 있다.

```python
fig, ax = plt.subplots()
ax.set_xticks([])
```

> - subplot을 1개만 그리면, 객체지향 방식으로 matplotlib을 쓸 수 있다.
